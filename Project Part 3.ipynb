```python
df1 = pd.read_csv('Proj1_1.csv')
df2 = pd.read_csv('Proj1_2.csv')
MergedDF = pd.merge(df1, df2, on='This_Week_Rank')
MergedDF = MergedDF.drop(['Unnamed: 8'],axis=1)
```

## THE ANALYSIS:
### Menu

```python
def runMenu(df):
    #Menu System
    quit = False
    while quit == False:
        print("\nMENU")
        print("1. Get Regression between Value of Streams and Total Revenue")
        print("2. Get AVONA test between Total Revenue and Value of Streams")
        print("3. Get Graph - Pie Chart of Value Counts of Genre")
        print("4. Output Value Counts of Genre based on if the song features anyone to an Excel file")
        print("5. View results from Testing for Equal Variances for 2 columns of your choice")
        print("6. View results from Comparing Means for 2 columns of your choice")
        print("7. View results from Regression of Certain Columns")
        print("8. Get Correlation Matrix")
        print("9. Quit")
        menu_choice = input("What is your choice: ")
        try:
            menu_choice = int(menu_choice) #convert to Integer
        except:
            print("ERROR : Please enter 1, 2, 3, 4, 5, 6, 7, 8 or 9")
            continue
        if menu_choice not in [1,2,3,4,5,6,7,8,9]:
            print("ERROR : Please enter 1, 2, 3, 4, 5, 6, 7, 8 or 9")
        else:
            
            if menu_choice==1:
                model1 = ols("Value_of_Streams ~ Total_Revenue", data=MergedDF).fit()
                xvar = 'Value_of_Streams'
                yvar = 'Total_Revenue'
                model = model1
                print('Coefficients: ') #Parameters
                print(model.params)
                intercept = model1.params[0]
                slope = model.params[1]
                print('Coeff X1: ', slope)
                r2 = model.rsquared
                print('R2: ', r2)
                print('F Test Statistic: ', model.fvalue)
                reg_p_val = model.f_pvalue
                print('p-value: ',reg_p_val)
                print('t Test Statistics: ')
                print(model.tvalues)
                print('t p-values: ')
                print(model.pvalues)
                if np.sign(slope) < 1:
                    slsign = ""
                else:
                    slsign = "+"
                regeq = f"{yvar} = {round(intercept,3)} {slsign} {round(slope,3)} {xvar}"
                print(regeq)
            
            if menu_choice==2:
                alpha = .05
                f, f_p_val = stats.f_oneway(MergedDF['Total_Revenue'],MergedDF['Value_of_Streams'])
                print("This is a test of equal means for TOTAL REVENUE and VALUE OF STREAMS with Ho: The means of all groups are equal/Ha: At least one group mean is different")
                print(f"The F test statistic is {round(f,3)} and the p-value is {round(f_p_val,4)}")
                if f_p_val < alpha:
                    print("Conclusion: Reject Ho: At least one group mean is different")
                    ANOVA_type = "ANOVA: At least one group mean different"
                else:
                    print("Conclusion: Fail to Reject Ho: We can't reject that the means are the same")
                    ANOVA_type = "ANOVA: Group Means are the same"
                print(ANOVA_type)
                y=[MergedDF['Total_Revenue'],MergedDF['Value_of_Streams']]
                plt.boxplot(y)
                plt.title(f'F: {round(f,3)}, p-val: {round(f_p_val,4)}',size=10)
                plt.suptitle(ANOVA_type,size=10)

                plt.xticks(range(1,3),[f"Total Revenue: {round(MergedDF['Total_Revenue'].mean(),2)}", f"Value of Stream: {round(MergedDF['Value_of_Streams'].mean(),2)}"])
                plt.xlabel('Tukey Result: Total Revenue differs from Value of Stream') #From below
                plt.ylabel('Range')
                plt.show()
            
            if menu_choice==3: #Takes time to load
                l=MergedDF.groupby('Music_Label').Change_percentage.sum()
                plt.plot(l.index, l)
                plt.title('Change in Percentage based on Music Label')
                plt.xlabel('Years')
                plt.ylabel('Adjusted Close Price')
                plt.xticks(rotation=90)
                plt.rcParams["figure.figsize"] = (20,1)
                plt.show()
            
            if menu_choice==4:
                xw.apps.add()
                wb = xw.Book()
                sht = wb.sheets['Sheet1']
                #df_SD=MergedDF.Genre.value_counts().groupby('Features_anyone').Quantity.count()
                df_SD = MergedDF.groupby(['Genre','Features_anyone']).size()
                sht.range('A1').value = df_SD
                chart = sht.charts.add(left=0, top=100, width=600, height=250)
                chart.set_source_data(sht.range('A1:C10'))
            
            if menu_choice==5:
                alpha = .05
                a = input("What is your first column? (This_Week_Rank, Last_Week_Rank, Length_of_Song, Total_Revenue, Change_percentage, Number_of_Sales, Amount_per_Sale, Number_of_Streams, Value_of_Streams)")
                b = input("What is your first column? (This_Week_Rank, Last_Week_Rank, Length_of_Song, Total_Revenue, Change_percentage, Number_of_Sales, Amount_per_Sale, Number_of_Streams, Value_of_Streams)")               
                t_var, p_val_var = stats.bartlett(MergedDF[a],MergedDF[b])
                print("This is a test of equal variances with Ho: The variances are equal")
                print(f"The t test statistic is {round(t_var,3)} and the p-value is {round(p_val_var,4)}")
                if p_val_var < alpha:
                    print(f"Conclusion: Reject Ho: The variances are not equal for {a} and (b)")
                    t_eq_var=False
                    t_type='Welch (unequal variances) Two-Sample t test'
                else:
                    print(f"Conclusion: Fail to Reject Ho: We can't reject that the variances are the same for {a} and (b)")
                    t_eq_var=True
                    t_type='Two-Sample t test (assuming equal variances)'
            
            if menu_choice==6:
                a = input("What is your first column? (This_Week_Rank, Last_Week_Rank, Length_of_Song, Total_Revenue, Change_percentage, Number_of_Sales, Amount_per_Sale, Number_of_Streams, Value_of_Streams)")
                b = input("What is your first column? (Track_Name, Artist, Music_Label, Genre, Features_anyone)")               
                from statsmodels.stats.multicomp import pairwise_tukeyhsd
                tukey = pairwise_tukeyhsd(endog=MergedDF[a], groups=MergedDF[b], alpha=0.05)
                print('Ho: The group means are equal')
                print(tukey.summary())
                tukey.plot_simultaneous()    # Plot group confidence intervals
                n_groups = len(MergedDF[b].unique()) #How many groups
                plt.vlines(x=MergedDF[a].mean(),ymin=-0.5,ymax=(n_groups-1)+.5, color="red")
                plt.show()
                
            if menu_choice==7:
                print("a. View Regression results from Total_Revenue and Genre")
                print("b. View Regression results from Total_Revenue and Genre and Change percentage")
                print("c. View Regression results from Total_Revenue and Genre and Change percentage and if it Features anyone")
                print("d. View Regression results from Value_of_Streams and Genre")
                print("e. View Regression results from Value_of_Streams and Genre and Change percentage")
                print("f. View Regression results from Value_of_Streams and Genre and Change percentage and if it Features anyone")
                print("g. Back to Main Menu")
                menu_choice1 = input("What is your choice: ")
                try:
                    menu_choice1 = menu_choice1
                except:
                    print("ERROR : Please enter a, b, c, d, e, f, g")
                    continue
                if menu_choice1 not in ['a','b','c','d','e','f','g']:
                    print("ERROR : Please enter a, b, c, d, e, f, or g")
                else:

                    if menu_choice1=='a':
                        model1 = ols("Total_Revenue ~ Genre", data=MergedDF).fit()
                        print(model1.summary())
                        
                    if menu_choice1=='b':
                        model2 = ols("Total_Revenue ~ Genre + Change_percentage", data=MergedDF).fit()
                        print(model2.summary())
                        
                    if menu_choice1=='c':
                        model3 = ols("Total_Revenue ~ Features_anyone + Genre + Change_percentage", data=MergedDF).fit()
                        print(model3.summary())
                        
                    if menu_choice1=='d':
                        model1 = ols("Value_of_Streams ~ Genre", data=MergedDF).fit()
                        print(model1.summary())
                        
                    if menu_choice1=='e':
                        model2 = ols("Value_of_Streams ~ Genre + Change_percentage", data=MergedDF).fit()
                        print(model2.summary())
                        
                    if menu_choice1=='f':
                        model3 = ols("Value_of_Streams ~ Features_anyone + Genre + Change_percentage", data=MergedDF).fit()
                        print(model3.summary())
                        
                    if menu_choice1=='g':
                        continue
            
            if menu_choice==8: # Corr matrix
                sns.pairplot(MergedDF)
                plt.show()
                
            if menu_choice==9:
                quit = True
df1 = pd.read_csv('Proj1_1.csv')
df2 = pd.read_csv('Proj1_2.csv')
MergedDF = pd.merge(df1, df2, on='This_Week_Rank')
MergedDF = MergedDF.drop(['Unnamed: 8'],axis=1)

runMenu(MergedDF)
                
```
